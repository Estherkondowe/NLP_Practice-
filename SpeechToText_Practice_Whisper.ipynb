```python
import whisper
# Load Whisper model
model = whisper.load_model("tiny")  # You can change to "tiny", "base", "medium", "large"
print("Model loaded!")
```

```python
audio_path = "/content/Shania Twain.mp3"
#change the path for your own file name
# Transcribe
result = model.transcribe(audio_path)
print("\nTranscription:")
print(result["text"])
```

```python
# Upload file
filename = "/content/abusa.wav"
# Load Chichewa Whisper model
# Use a pipeline as a high-level helper
from transformers import pipeline

pipe = pipeline("automatic-speech-recognition", model="dmusingu/WHISPER-SMALL-SWAHILI-ASR-CV-14")
# Transcribe
result = pipe(filename)
print("\nTranscription:")
print(result["text"])
```
