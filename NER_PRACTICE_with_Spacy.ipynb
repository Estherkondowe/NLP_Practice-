```python
!pip install spacy
# Download the English language model for spaCy
!python -m spacy download en_core_web_sm

import spacy
# Load spaCy model
nlp = spacy.load("en_core_web_sm")
```

```python
# Example article URL

urlA="https://www.nbcnews.com/health/health-news/flu-rsv-covid-cicada-virus-2026-symptoms-signs-rcna265906"
url6="https://www.code-blossom.com/blog/michelles-story"


text = fetch_article_text(urlA)
print(f"Fetched {len(text)} characters from the article.")

# Process the text with spaCy
doc = nlp(text)

# Print the Named Entities
for ent in doc.ents:
    print(f"Entity: {ent.text}, Label: {ent.label_}")
```

```python
sentence = "Flu and Covid, including a new variant called BA.3.2, nicknamed “cicada,” are still circulating, along with several other respiratory illnesses and a nasty stomach bug."
doc_sentence = nlp(sentence)

print("Sentence Dependencies:")
for token in doc_sentence:
    print(f"Token: {token.text}, Dependency: {token.dep_}, Head: {token.head.text}, Head POS: {token.head.pos_}, Children: {[child.text for child in token.children]}")
```

```python
from spacy import displacy

# Visualize the dependency tree
displacy.render(doc_sentence, style="dep", jupyter=True, options={'compact': True, 'distance': 90})
```
